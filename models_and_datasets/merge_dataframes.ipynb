```python
import pandas as pd

dataset_terpenoids_phenolics = pd.read_csv("phenolics_and_terpenoids.csv")
```

```python
dataset_terpenoids_phenolics
```

```python
#get data with only zeros in the last 23 columns
zero_columns = dataset_terpenoids_phenolics.columns[(dataset_terpenoids_phenolics == 0).all()]

print(zero_columns)
```

```python
dataset_alkaloid = pd.read_csv("alkaloid_data_processed.csv")
# dataset_alkaloid.head()
```

```python
dataset_alkaloid.drop(columns=["Name", "PM"], inplace=True)
```

```python
dataset_alkaloid.columns = ["ids", "smiles"] + list(dataset_alkaloid.columns[2:])
```

```python
dataset_alkaloid
```

```python
# merge the two datasets
dataset = pd.concat([dataset_terpenoids_phenolics, dataset_alkaloid], ignore_index=True)
```

```python
dataset.fillna(0, inplace=True)
```

```python
zero_columns = dataset.columns[(dataset.iloc[2:] == 0).all()]

print(zero_columns)
```

```python
zero_rows = dataset.iloc[2:].index[(dataset.iloc[2:] == 0).all(axis=1)]

print(zero_rows)
```

```python
dataset.to_csv("final_dataset_top_v2.csv", index=False)
```
